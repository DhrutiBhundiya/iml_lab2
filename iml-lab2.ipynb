```python
### • to append values at the end of an array
#Dhrti bhundiya 
# 216270307016
import numpy as np
x = [10, 20, 30]
print("Original array:")
print(x)
x = np.append(x, [40])
print("After append values to the end of the array:")
print(x)
```

```python
### • to create another shape from an array without changing its data (3*2 to 2*3)
#Dhrti bhundiya 
# 216270307016

import numpy as np
x = np.array([1, 2, 3, 4, 5, 6])
y = np.reshape(x,(3,2))
print("Reshape 3x2:")
print(y)
z = np.reshape(x,(2,3))
print("Reshape 2x3:")
print(z)
```
